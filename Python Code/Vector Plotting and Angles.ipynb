```python
#this is inner product to find angle in radians
import numpy as np
x = np.matrix([[1],[1]])
y = np.matrix([[-1],[1]])
c = np.matrix([[2,-1],[-1,4]])

np.arccos((x.T @ c @ y) / np.sqrt((x.T @ c @ x) @ (y.T @ c @ y)))

```

```python
import numpy as np
x = np.matrix([[1],[1],[1]])
y = np.matrix([[2],[-1],[0]])
c = np.matrix([[1,0,0],[0,2,-1],[0,-1,3]])

np.arccos((x.T @ c @ y) / np.sqrt((x.T @ c @ x) @ (y.T @ c @ y)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

Va = np.array([2,3])
Vb = np.array([4,1])
Vc = np.array(Va**2)


V = np.array([Va,Vb,Vc])
origin = [0], [0] # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
M = np.array([[2,3],[4,1],[1,1]])

rows,cols = M.T.shape

#Get absolute maxes for axis ranges to center origin
#This is optional
maxes = 2*np.amax(abs(M), axis = 0)
colors = ['b','r','k']


for i,l in enumerate(range(0,cols)):
    plt.axes().arrow(0,0,M[i,0],M[i,1],head_width=0.05,head_length=0.1,color = colors[i])

plt.plot(0,0,'ok') #<-- plot a black point at the origin
plt.axis('equal')  #<-- set the axes to the same scale
plt.xlim([-maxes[0],maxes[0]]) #<-- set the x axis limits
plt.ylim([-maxes[1],maxes[1]]) #<-- set the y axis limits
plt.grid(b=True, which='major') #<-- plot grid lines
plt.show()
```
